```python
import pandas as pd
```

```python
df= pd.read_csv('data_linear.csv')
df
```

```python
X = df[['age', 'bmi', 'children']]
y = df['charges']
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)
ols_model = sm.OLS(y, X)
fitted_model = ols_model.fit()
fitted_model.params
fitted_model.summary()
```

```python
x = df[['age', 'bmi']]
```

```python
x = sm.add_constant(x)
ols_model = sm.OLS(y, x)
fitted_model = ols_model.fit()
fitted_model.params
fitted_model.summary()
```
